### principaux testes testes associé a une régression lineéaire

The answer to this question lies in testing for the assumptions underlying the statistical bases for multivariate analysis. We already did some data cleaning and discovered a lot about 'SalePrice'. Now it's time to go deep and understand how 'SalePrice' complies with the statistical assumptions that enables us to apply multivariate techniques.

According to Hair et al. (2013), four assumptions should be tested:

Normality - When we talk about normality what we mean is that the data should look like a normal distribution. This is important because several statistic tests rely on this (e.g. t-statistics). In this exercise we'll just check univariate normality for 'SalePrice' (which is a limited approach). Remember that univariate normality doesn't ensure multivariate normality (which is what we would like to have), but it helps. Another detail to take into account is that in big samples (>200 observations) normality is not such an issue. However, if we solve normality, we avoid a lot of other problems (e.g. heteroscedacity) so that's the main reason why we are doing this analysis.

Homoscedasticity - I just hope I wrote it right. Homoscedasticity refers to the 'assumption that dependent variable(s) exhibit equal levels of variance across the range of predictor variable(s)' (Hair et al., 2013). Homoscedasticity is desirable because we want the error term to be the same across all values of the independent variables.

Linearity- The most common way to assess linearity is to examine scatter plots and search for linear patterns. If patterns are not linear, it would be worthwhile to explore data transformations. However, we'll not get into this because most of the scatter plots we've seen appear to have linear relationships.

Absence of correlated errors - Correlated errors, like the definition suggests, happen when one error is correlated to another. For instance, if one positive error makes a negative error systematically, it means that there's a relationship between these variables. This occurs often in time series, where some patterns are time related. We'll also not get into this. However, if you detect something, try to add a variable that can explain the effect you're getting. That's the most common solution for correlated errors.

#### Normalité

Le point ici est de tracer la distribution   de la variable 'SalePrice' pour avoir une vue sur sa normalité et son asymétrie.

```python
sns.distplot(train['SalePrice'] , fit=norm);

#Récupèrer les paramètres ajustés utilisés par la fonction
(mu, sigma) = norm.fit(train['SalePrice'])
#Tracer la ditribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequence')
plt.title('Distribution de SalePrice')
```

En statistiques, le diagramme Quantile-Quantile ou diagramme Q-Q ou Q-Q plot est un outil graphique permettant d'évaluer la pertinence de l'ajustement d'une distribution donnée à un modèle théorique.

```python
#Diagramme Quantile-Quantile
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

L'ajustement par une loi normale n'est pas toujours légitime, ainsi dans le diagramme ci-contre la distribution des quantiles observés en comparaison avec les quantiles d'une loi normale laisse penser que la distribution suit davantage une loi exponentielle

```python

fig = plt.figure()
res = stats.probplot(train['SalePrice'],dist = stats.expon, plot=plt)
plt.show()
```

Aussi La variable cible est tournée à  la droite(Asymétrie). Comme les modèles (linéaires) aiment les données normalement distribuées, nous devons transformer cette variable et la rendre plus normalement distribuée.

#### La transformation Log de la variable cible

```python
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
#On utilise la fonction Lop1p de numpy qui applique log(1+x) a tous les ellements de la collone
train["SalePrice"] = np.log1p(train["SalePrice"])

#Vérifier la nouvelle distribution 
sns.distplot(train['SalePrice'] , fit=norm);

# Récupèrer les paramètres ajustés utilisés par la fonction
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Visualiser la distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#et la diagramme Q-Q
fig = plt.figure()
res = stats.probplot(train['SalePrice'],plot=plt)
plt.show()
```

Le asymétrie semble maintenant corrigé et les données semblent plus normalement distribuées.

#### Homoscedasticité

La meilleure approche pour tester l'homoscédasticité pour deux variables métriques est graphiquement. Les départs à partir d'une dispersion égale sont représentés par des formes telles que des cônes (petite dispersion d'un côté du graphique, grande dispersion du côté opposé) ou des diamants (un grand nombre de points au centre de la distribution).

## Définition de la stratégie de Validation croisée

La validation croisée (« cross-validation ») est, en apprentissage automatique, une méthode d’estimation de fiabilité d’un modèle fondé sur une technique d’échantillonnage.

Nous utilisons la fonction cross_val_score de Sklearn. Cependant cette fonction n'a pas d'attribut 'mélanger', nous ajoutons alors une ligne de code, afin de mélanger l'ensemble de données avant la validation croisée

```python
'''
Nous utilisons la fonction cross_val_score de Sklearn. Cependant cette fonction n'a pas d'attribut 'mélanger', 
nous ajoutons alors une ligne de code, afin de mélanger l'ensemble de données avant la validation croisée
'''

#Fonction de Validation
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(train.values)
    rmse= np.sqrt(-cross_val_score(model, train.values, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

## Analyse de covariance

L’analyse de covariance est un outil extrêmement  utile pour l’analyse de données. Dans le modèle linéaire Y ~ X1,X2,...,Xp quand certains des variables explicatives sont  quantitatives et d’autres sont qualitatives ce modèle s’appelle l’analyse de covariance.
Notons que les variables qualitatives ne fournissent pas autant d'informations que les variables quantitatives. 
<br>
Pour illustrer le rôle de la covariable dans l'ANCOVA, examinons une situation dans laquelle les enquêteurs comparent les salaires annuels des diplômés universitaires masculins et féminins. Un échantillon aléatoire de 5 individus pour chaque sexe est compilé, et une ANOVA simple à un facteur est effectuée:


```python
data = pd.DataFrame(columns = ['sex','salary','years'])
col =  ['sex','salary','years']
L= []
L.append(pd.DataFrame( [['m' ,78 ,3]] , columns=col))
L.append(pd.DataFrame( [[ 'm', 43, 1]] , columns=col))
L.append(pd.DataFrame([ ['m', 103, 5 ]] , columns=col))
L.append(pd.DataFrame( [[ 'm' ,48 ,2] ], columns=col))
L.append(pd.DataFrame( [['m', 80, 4 ]] , columns=col))
L.append(pd.DataFrame([ ['f', 80, 5 ]] , columns=col))
L.append(pd.DataFrame( [[ 'f', 50, 3]] , columns=col))
L.append(pd.DataFrame( [[ 'f', 30, 2]] , columns=col))
L.append(pd.DataFrame( [[ 'f', 20, 1]] , columns=col))
L.append(pd.DataFrame( [[ 'f', 60, 4]] , columns=col))

df =pd.concat(L)
```

```python
df[['sex','salary']]
```

```python
import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")
from statsmodels.formula.api import ols
formula = 'salary ~ sex'  
lm = ols(formula,df[['sex','salary']])
fit = lm.fit()
fit.summary()
```

### Etapes de l'ANCOVA:

Deuxièmement, nous devons être sûrs que la relation de régression de la réponse avec la covariable a la même pente pour chaque groupe de traitement. C'est un point extrêmement important. Dans notre exemple, nous devons nous assurer que les lignes pour les mâles et les femelles sont parallèles (ont une pente égale).

Selon le résultat de l'essai pour des pentes égales, nous avons deux manières alternatives de terminer l'ANCOVA: 1) adapter un modèle de pente commun  ou 2) évaluer les différences de moyennes à au moins trois niveaux de la covariable.

```python
train['SalePrice']=y_train
formula = 'SalePrice ~GrLivArea *C(HasBsmt) '  # ANCOVA formula
lm = ols(formula, train)
fit = lm.fit()
fit.summary()
```

```python
fig = plt.figure()
fig, ax = plt.subplots()
train = train.sort_values('GrLivArea')
ax.scatter(train['GrLivArea'][train['HasBsmt'] ==train['HasBsmt'].unique()[1]], train['SalePrice'][train['HasBsmt'] ==train['HasBsmt'].unique()[1]],label='Avec sous-sol')
ax.scatter(train['GrLivArea'][train['HasBsmt'] ==train['HasBsmt'].unique()[0]], train['GrLivArea'][train['HasBsmt'] ==train['HasBsmt'].unique()[0]], color  = 'r', label='Sans sous-sol')
ax.plot(train['GrLivArea'][train['HasBsmt'] ==train['HasBsmt'].unique()[1]], train['GrLivArea'][train['HasBsmt'] ==train['HasBsmt'].unique()[1]].apply(lambda x: 8.1036 +0.2980*x),label='Avec sous-sol')
ax.plot(train['GrLivArea'][train['HasBsmt'] ==train['HasBsmt'].unique()[0]], train['GrLivArea'][train['HasBsmt'] ==train['HasBsmt'].unique()[0]].apply(lambda x: 8.7109 +0.2219*x ) , color  = 'r', label='Sans sous-sol')
plt.ylabel('SalePrice', fontsize=13)
plt.xlabel('GrLivArea', fontsize=13)
ax.legend()
plt.show()
```

Malgré que nous voyons  une différence significative à chaque niveau de la covariable spécifié 'HasBsmt'et que l'empleur des prix des propriétés entre celles ayant un sous-sol et celles qui ne l'ont pas diffère (donnant lieu à la signification de l'interaction). Les résultats de l'ANCOVA nous  donne un p-value=0.07 >0.05 pour le modèle SalePrice ~ GrLivArea*C(HasBsmt), donc on ne rejète pas H0, cependant ceci peut-etre au manque de donnée concernant les propriétés sans sous-sol (1421 ayant un sous-sol contre 37 n'ayant pas de sous-sol dans notre base de données) ou à l'existance d'autre multicolineéarité dans notre jeu de données.
